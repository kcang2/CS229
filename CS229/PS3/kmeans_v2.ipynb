```python
from matplotlib.image import imread; import matplotlib.pyplot as plt; import numpy as np;
```

```python
large = imread('mandrill-large.tiff')
```

```python
small = imread('mandrill-small.tiff')
```

```python
A = small.reshape(-1,1,3).astype(np.int32)
```

```python
miu = A[np.random.randint(0,len(A),16)].reshape(-1,3)
```

```python
j=0;
```

```python
while j<30:
    C=np.argmin(np.sum((A-miu)**2,axis=2),axis=1)
    for i in range(16):
        miu[i]=np.mean(A[C==i].reshape(-1,3),axis=0)
    j=j+1
```

```python
L=large.reshape(-1,1,3).astype(np.int32)
```

```python
compressed=miu[np.argmin(((L-miu)**2).sum(axis=2),axis=1)]
```

```python
plt.imshow(compressed.reshape(large.shape).astype(np.uint8))
```
