```python
import numpy as np
```

```python
import scipy.spatial.distance as sd 
```

```python
from matplotlib.image import imread; import matplotlib.pyplot as plt;
```

```python
large = imread('mandrill-large.tiff')
```

```python
plt.imshow(large)
```

```python
small=imread('mandrill-small.tiff')
```

```python
A=np.reshape(small,[-1,3])
```

```python
miu=A[np.random.choice(np.arange(len(A)), 16, False)]
```

```python
J=0; Jprev=10e5; j=0; c=np.empty(shape=(16,) + A.shape[1:])
```

```python
while(np.absolute(J-Jprev) and j<30):
    dist=sd.cdist(A,miu,'sqeuclidean')
    c=np.argmin(dist,axis=1)
    for i in range(16):
        miu[i]=np.mean(A[c==i],axis=0)
    j=j+1
    
        
```

```python
L=np.reshape(large,[-1,3])
```

```python
L=miu[np.argmin(sd.cdist(L,miu,'sqeuclidean'),axis=1)][:]
```

```python
compress=L.reshape(large.shape[0],large.shape[1],large.shape[2])
```
